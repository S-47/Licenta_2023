```python
knn=KNeighborsClassifier(n_neighbors=9, algorithm='auto', metric='euclidean', weights='distance')
knn.fit(x_train, y_train)
y_pred_knn=knn.predict(x_test)

cm = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(6,5))
p = sns.heatmap(cm, annot=True, cmap="BuPu", fmt='g')
plt.title('Matricea de confuzie pentru KNN')
plt.xlabel('Clasa estimată cu KNN' )
plt.ylabel('Clasa reală')
plt.show()
```

```python
acc_score_knn=round(accuracy_score(y_test, y_pred_knn),2)
jaccard_score_knn = round(jaccard_score(y_test , y_pred_knn, average='micro'),2)
f1_score_knn=round(f1_score(y_test, y_pred_knn, average='micro'),2)
precision_score_knn=round(precision_score(y_test , y_pred_knn),2)
recall_score_knn=round(recall_score(y_test , y_pred_knn),2)

plt.figure(figsize=(5,3))
scores = np.array([[acc_score_knn, precision_score_knn, recall_score_knn, f1_score_knn, jaccard_score_knn]])
sns.heatmap(scores.T, cmap='BuPu', annot=True, fmt='g', xticklabels=['KNN'], yticklabels=['Acuratețe', 'Precizie', 'Sensibilitate', 'F1', 'Jaccard'])
plt.title('Rezultate')
plt.show()
```

```python
model_accuracy['KNeighborsClassifier']=acc_score_knn
model_recall['KNeighborsClassifier']=recall_score_knn
model_precision['KNeighborsClassifier']=precision_score_knn
model_jaccard['KNeighborsClassifier']=jaccard_score_knn
model_f1['KNeighborsClassifier']=f1_score_knn
```

```python
x_new = df.drop("Status", axis=1)
y_new= df["Status"]
```

```python
x_train_over, x_test_over, y_train_over, y_test_over= train_test_split(x_new, y_new, test_size=0.2, random_state=42)

model_recall={}
model_auc={}
model_accuracy={}
model_precision={}
model_jaccard={}
```

```python
warnings.filterwarnings("ignore")

LR=LogisticRegression()


pipe_lr= Pipeline ([
    ('oversampling', SMOTE()),
    ('LR', LR)
])

C_list_lr=[0.001, 0.01, 0.5, 0.1, 1, 10, 100 ]
solver_list_lr=['saga']
penalty_list_lr=[None, 'elasticnet', 'l1']
max_iter_list_lr=[100, 1000, 3000]
random_state_list_lr=[None, 20, 42 ]
param_grid_lr = {
    'LR__C': C_list_lr,
    'LR__solver': solver_list_lr,
    'LR__penalty': penalty_list_lr,
    'LR__max_iter': max_iter_list_lr,
    'LR__random_state': random_state_list_lr
}


grid_lr = GridSearchCV(pipe_lr, param_grid_lr, cv=5, scoring='balanced_accuracy', return_train_score=False)
grid_lr.fit(x_train_over, y_train_over)
```

```python
print(grid_lr.best_params_)
```

```python
prediction_lr=grid_lr.predict(x_test_over)
grid_lr_matrix = confusion_matrix(y_test_over, prediction_lr)

plt.figure(figsize=(6,5))
p = sns.heatmap(grid_lr_matrix, annot=True, cmap="BuPu", fmt='g')
plt.title('Matricea de confuzie pentru Logistic Regression ')
plt.xlabel('Clasa estimată cu LR' )
plt.ylabel('Clasa reală')
plt.show()
```

```python
acc_score_lr=round(accuracy_score(y_test_over, prediction_lr),2)
jaccard_score_lr=round(jaccard_score(y_test_over , prediction_lr),2)
f1_score_lr=round(f1_score(y_test_over, prediction_lr),2)
precision_score_lr=round(precision_score(y_test_over , prediction_lr),2)
recall_score_lr=round(recall_score(y_test_over , prediction_lr),2)

plt.figure(figsize=(5,3))
scores = np.array([[acc_score_lr, precision_score_lr, recall_score_lr, f1_score_lr, jaccard_score_lr]])
sns.heatmap(scores.T, cmap='BuPu', annot=True, fmt='g', xticklabels=['LR'], yticklabels=['Acuratețe', 'Precizie', 'Sensibilitate', 'F1', 'Jaccard'])
plt.title('Rezultate')
plt.show()
```

```python
prediction_prob_lr = grid_lr.predict_proba(x_test_over)[:,1]
false_positive_lr, true_positive_lr, threshold1 = roc_curve(y_test_over, prediction_prob_lr)
auc_lr = roc_auc_score(y_test_over, prediction_prob_lr)
plt.figure(figsize=(5,3))
plt.plot(false_positive_lr, true_positive_lr,label="AUC="+str(auc_lr))
plt.ylabel('Specificitatea')
plt.xlabel('Sensibilitatea')
plt.title("Curba ROC pentru LR ")
plt.legend(loc=5)
plt.show()
```
